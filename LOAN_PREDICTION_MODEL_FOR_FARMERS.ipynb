**LOAN PREDICTION MODEL FOR FARMERS IN STATES IN INDIA**
The loan_data_final table is a combintion of a weather data that takes into account the Total Replenishable groundwater resource per state for the year 2003, a remote sensing data of the number of fire points communicated by the forest survey of india by remote sensing for the year 2014, total yield numbers for maize by state for the year 2014 and a state-wise agricultural loan disbursement to farmers by commercial banks. I manually generated the loan status randomly.
I obtained most of the data from https://data.gov.in and www.indianwaterportal.org

```python
import pandas as pd
```

```python
yield_data = pd.read_csv('Table_8.4-Statewise_1.csv')
```

```python
yield_data = yield_data[['State/ UT Name','Foodgrains(cereals) - Maize (Total) - 2014-15']]
```

```python
yield_data
```

```python
climate_data = pd.read_csv('Water_resource_file_2013.csv',encoding='cp1252')
```

```python
climate_data = climate_data[['State/ Union Territory','Total Replenishable groundwater resource']]
```

```python
climate_data
```

```python
climate_data.rename(columns= {'State/ Union Territory': 'State/ UT Name'}, inplace=True)
```

```python
sensing_data = pd.read_csv('Rajya_Sabha_Session_234_AU2438_1.csv')
```

```python
sensing_data = sensing_data[['Name of States/UTs','2014']]
```

```python
sensing_data
```

```python
sensing_data.rename(columns= {'Name of States/UTs': 'State/ UT Name'}, inplace=True)
```

```python
sensing_data
```

```python
loan_per_state_data = pd.read_csv('rs_session-242_au2716_1.1.csv')
```

```python
loan_data = yield_data.merge(climate_data, on='State/ UT Name', how='inner')
```

```python
loan_data
```

```python
loan_data_2 = loan_data.merge(sensing_data, on='State/ UT Name', how='inner')
```

```python
loan_data_2
```

```python
loan_data_final = loan_data_2.merge(loan_per_state_data, on='State/ UT Name', how='inner')
```

```python
loan_data_final
```

```python
loan_data_final.rename(columns= {'2014': 'Remote Sensing'}, inplace=True)
```

```python
loan_data_final
```

```python
loan_data_final = loan_data_final.drop(['Sl. No.'],axis=1)
```

```python
loan_data_final
```

```python
import numpy as np
loan_data_final['loan status'] = np.random.choice([0,1], len(loan_data_final))
```

```python
loan_data_final
```

```python
loan_data_final.describe()
```

```python
loan_data_final.info()
```

```python
loan_data_final.shape
```

```python
loan_data_final.isnull().any()
```

```python
loan_data_final.isnull().sum()
```

```python
loan_data_final['Foodgrains(cereals) - Maize (Total) - 2014-15'].fillna(loan_data_final['Foodgrains(cereals) - Maize (Total) - 2014-15'].mean(),inplace = True)
```

```python
loan_data_final['Number of Application collected'].fillna(loan_data_final['Number of Application collected'].mean(),inplace = True)
```

```python
loan_data_final['Numbers of Applications given to Banks'].fillna(loan_data_final['Numbers of Applications given to Banks'].mean(),inplace = True)
```

```python
loan_data_final['Numbers of applications sanctioned by banks'].fillna(loan_data_final['Numbers of applications sanctioned by banks'].mean(),inplace = True)
```

```python
loan_data_final['Number of RuPay Card issued'].fillna(loan_data_final['Number of RuPay Card issued'].mean(),inplace = True)
```

```python
loan_data_final['Total Amount sanctioned (Rs. in Lakh)'].fillna(loan_data_final['Total Amount sanctioned (Rs. in Lakh)'].mean(),inplace = True)
```

```python
loan_data_final['Total amount disbursed (Rs. in Lakh)'].fillna(loan_data_final['Total amount disbursed (Rs. in Lakh)'].mean(),inplace = True)
```

```python
loan_data_final.info()
```

```python
loan_data_final.loc[(loan_data_final['Total Replenishable groundwater resource']=='1.40*')] = 1.40
```

```python
loan_data_final.loc[(loan_data_final['Total Replenishable groundwater resource']=='0.07*')] = 0.07
```

```python
loan_data_final['loan status'] = loan_data_final['loan status'].astype(int)
```

```python
loan_data_final
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
X=loan_data_final[['Foodgrains(cereals) - Maize (Total) - 2014-15','Number of Application collected','Numbers of Applications given to Banks','Numbers of applications sanctioned by banks','Number of RuPay Card issued',
        'Total Amount sanctioned (Rs. in Lakh)','Total amount disbursed (Rs. in Lakh)','Remote Sensing','Total Replenishable groundwater resource']].values
y=loan_data_final["loan status"].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
expected = y_test
predicted = model.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(expected, predicted))
```

```python
print(metrics.confusion_matrix(expected, predicted))
```
